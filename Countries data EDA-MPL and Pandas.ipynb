#### Q2. What is the distribution of life expectancy across all countries and years?


```python
plt.hist(countries['lifeExpectancy'],facecolor="skyblue",edgecolor="blue",bins=20)
plt.xlabel("Life Expectancy")
plt.ylabel("Frequency")
plt.title("Histogram Plot of Life Expectancy")
plt.show()
```

#### Q3. How does life expectancy vary over the years for a specific country (e.g., Afghanistan)?
https://www.geeksforgeeks.org/matplotlib-markers-module-in-python/
https://www.geeksforgeeks.org/grids-in-matplotlib/


```python
countriesafg=countries[countries['country']=='Afghanistan']
countriesafg
```

```python
fig,ax1=plt.subplots()
fig=plt.figure(figsize=(10,6))
ax1.set_xlabel("Year")
ax1.tick_params(axis="x",bottom=False,labelbottom=True)
ax1.set_ylabel("Life Expectancy",color="Black",size="large")
ax1.set_title("Life Expectancy over the years in afghanistan")
ax1.tick_params(axis="y",labelcolor='black',labelsize="large")
ax1.plot(countriesafg['year'],countriesafg['lifeExpectancy'],color="green",marker=".")
ax1.grid(True)
```

#### Q4. What is the relationship between GDP per capita and life expectancy?


```python
plt.scatter(countries['gdpPerCapita'],countries['lifeExpectancy'],color='red')
plt.xlabel("GDP per Capita ($)")
plt.ylabel("Life Expectancy")
plt.title("GDP per Capita vs. Life Expectancy")
plt.grid(True)
plt.show()
```

#### Q5. How has the population changed over time for a specific continent (e.g., Africa)?

** Use groupby for year with populations

```python
countriesaf=countries[countries['continent']=='Africa']
af=countriesaf.groupby(['year'])['population'].sum()
plt.plot(af,color='red',marker='.')
plt.grid()
plt.show()
```

#### Q6. What is the trend of GDP per capita over the years for a specific continent (e.g., Europe)?


```python
countrieseup=countries[countries['continent']=='Europe']
eup=countrieseup.groupby(['year'])['gdpPerCapita'].sum()
plt.plot(eup,color='blue',marker='.')
plt.grid()
plt.show()
```
